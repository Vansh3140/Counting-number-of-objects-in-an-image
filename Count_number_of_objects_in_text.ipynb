```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import cvlib as cv
from cvlib.object_detection import draw_bbox
from numpy.lib.polynomial import poly 
from google.colab.patches import cv2_imshow
```

```python
img1=cv2.imread('/content/cars.jfif')
img2=cv2.imread('/content/chaos.jfif')
img3=cv2.imread('/content/people.jfif')
```

```python
cv2_imshow(img1)
```

```python
cv2_imshow(img2)
```

```python
cv2_imshow(img3)
```

```python
box1,label1,count1=cv.detect_common_objects(img1)
box2,label2,count2=cv.detect_common_objects(img2)
box3,label3,count3=cv.detect_common_objects(img3)
```

```python
output1=draw_bbox(img1,box1,label1,count1)
output2=draw_bbox(img2,box2,label2,count2)
output3=draw_bbox(img3,box3,label3,count3)
```

```python
cv2_imshow(output1)
```

```python
cv2_imshow(output2)
```

```python
cv2_imshow(output3)
```

```python
print('Number of objects in Figure 1 are :' +str(len(label1)))
print('Number of objects in Figure 2 are :' +str(len(label2)))
print('Number of objects in Figure 3 are :' +str(len(label3)))
```
